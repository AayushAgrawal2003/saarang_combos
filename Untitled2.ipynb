```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel("./ONLINE ENGAGEMENT TASK 1.xlsx")
```

```python
df.head()
```

```python
current_series = pd.Series(['t'])
db = pd.DataFrame()
titles = [0]
for i in df:
    if i.split()[0] != "Unnamed:":
        titles.append(i)
        current_series.reset_index(drop=True)
        db = pd.concat([db,current_series],axis =1)
        #print(current_series)
        current_series = pd.Series()
        current_series = pd.concat([current_series,df[i]],axis = 0)
        current_series = current_series.reset_index(drop=True)        
    if i.split()[0] == "Unnamed:":
        current_series = pd.concat([current_series,df[i]],axis = 0)
        current_series = current_series.reset_index(drop=True)
        
db = pd.concat([db,current_series],axis =1)
```

```python
db.columns = titles
db = db.drop([0],axis =1)
db = db.drop([0],axis =0)
```
