```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
import csv
```

```python
df = pd.read_excel("./ONLINE ENGAGEMENT TASK 1.xlsx")
```

```python
df.head()
```

```python
current_series = pd.Series(['t'])
db = pd.DataFrame()
titles = [0]
for i in df:
    if i.split()[0] != "Unnamed:":
        titles.append(i)
        current_series.reset_index(drop=True)
        db = pd.concat([db,current_series],axis =1)
        #print(current_series)
        current_series = pd.Series()
        current_series = pd.concat([current_series,df[i]],axis = 0)
        current_series = current_series.reset_index(drop=True)        
    if i.split()[0] == "Unnamed:":
        current_series = pd.concat([current_series,df[i]],axis = 0)
        current_series = current_series.reset_index(drop=True)
        
db = pd.concat([db,current_series],axis =1)
```

```python
db.columns = titles
db = db.drop([0],axis =1)
db = db.drop([0],axis =0)
```

```python
db
```

```python
names = pd.unique(db.values.ravel())
```

```python
names = names[np.logical_not(pd.isnull(names))]
```

```python
names
```

```python
db1 = pd.DataFrame(columns = titles[1:])
db1
```

```python
master = []
for i in names:
    clubs = []
    for j in titles[1:]:
        if db[j].str.contains(i).sum() >= 1:
            clubs.append(j)
    #print(clubs)
    clubs = sorted(clubs)
    master.append(clubs)
    
master_new = []
for i in master:
    if len(i) > 1:
        master_new.append("_".join(i))
            
```

```python
master_new = dict(Counter(master_new))
```

```python
s = dict(sorted(master_new.items(), key=lambda x:x[1] , reverse = True))
```

```python
plt.figure(figsize =(20,5))
keys = list(s.keys())[:10]
vals = [s[k] for k in keys][:10]
sns.barplot(x=keys, y=vals)
```

```python
with open('combos.csv', 'w', newline='') as f:
    writer = csv.DictWriter(f, fieldnames=s.keys())
    writer.writeheader()
    writer.writerow(s)
```
